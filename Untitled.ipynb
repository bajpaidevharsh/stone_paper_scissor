```python
import cv2
```

```python
import time
```

```python
from ultralytics import YOLO
```

```python
import random
```

```python
model=YOLO('best.pt')
```

```python
score_user=0
score_computer=0
def score(user,comp):
    if comp=='stone':
        if user=='stone':
            comp='draw'
            user='draw'
        elif user=='paper':
            comp='lose'
            user='win'
        elif user=='scissor':
            comp='win'
            user='lose'
    elif comp=='paper':
        if user=='stone':
            comp='win'
            user='lose'
        elif user=='paper':
            comp='draw'
            user='draw'
        elif user=='scissor':
            comp='lose'
            user='win'
    elif comp=='scissor':
        if user=='stone':
            comp='lose'
            user='win'
        elif user=='paper':
            comp='win'
            user='lose'
        elif user=='scissor':
            comp='draw'
            user='draw'
    return(comp,user)
```

```python
cap = cv2.VideoCapture(0)
pp=['paper','scissor','stone']
win='neutral'
while True:
    success, frame = cap.read()
    if success:
        frame = cv2.resize(frame, (640, 480))
        frame = cv2.flip(frame, 1)
        results = model(frame, stream=True)
        cv2.putText(frame,'YOU',(10,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
        cv2.putText(frame,'COMPUTER',(450,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
        cv2.putText(frame,'win',(220,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
        cv2.putText(frame,'Q to exit',(450,450),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
        for result in results:
            boxes = result.boxes
            for box in boxes:
                conf=box.conf
                conf=float(conf)*100
                if conf>=35:
                    x1, y1, x2, y2 = box.xyxy[0]
                    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                    cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                    label=box.cls
                    label=int(label)
                    cv2.putText(frame, pp[label], (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
                    if conf>=45:
                        computer_choice=random.choice(pp)
                        kp=score(pp[label],computer_choice)
                        cv2.putText(frame,f'{pp[label]}',(10,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
                        cv2.putText(frame,f'{computer_choice}',(450,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),3,cv2.LINE_AA)
                        computer,user=kp
                        if computer=='win':
                            cv2.putText(frame,'COMP',(220,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3,cv2.LINE_AA)
                            time.sleep(2)
                            break
                        elif user=='win':
                            cv2.putText(frame,'YOU',(220,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3,cv2.LINE_AA)
                            time.sleep(2)
                            break
                        else:
                            cv2.putText(frame,'DRAW',(220,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3,cv2.LINE_AA)
                            time.sleep(2)
                            break
        cv2.imshow('output', frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
